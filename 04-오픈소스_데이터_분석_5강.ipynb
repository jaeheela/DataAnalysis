```python
import requests
import pandas as pd

# 서울시 에코마일리지 에너지사용량 통계정보(회원유형별)
api_key='[발급받은 인증키]'
url = 'http://openapi.seoul.go.kr:8088/'+api_key+'/json/energyUseDataSummaryInfo/1/5/'

params = {
    'KEY': api_key,
    'TYPE': 'json',
    'SERVICE': 'energyUseDataSummaryInfo',
    'START_INDEX': 1,
    'END_INDEX': 10
}

## 데이터 수집
response = requests.get(url, params=params)

## 호출 성공/실패 출력
print(response.json())
```

## 데이터 프레임 생성

```python
import pandas as pd

## requests로부터 response, body, item 항목 읽기
## requests로부터 energyUseDataSummaryInfo, row 항목 읽기
data = response.json()['energyUseDataSummaryInfo']['row']
print(type(data))
print(len(data))
print(data[0])
print(data[1])
print(data[2])
print(data[3])
print(data[4])
df = pd.DataFrame(data)
# df.head()
# df['YEAR']
df['MON']
```
