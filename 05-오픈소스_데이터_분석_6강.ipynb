### 6-1 데이터 훑어보기

```python
import pandas as pd
import numpy as np

data = {
    'name': ['김민수', '이지영', '박준호', '최서연', '정도윤'],
    'age': [25, 30, 28, 22, 35],
    'city': ['서울', '부산', '인천', '서울', '대전'],
    'score': [90, 85, 95, 80, np.nan]
}
df = pd.DataFrame(data)
```

```python
# 데이터 측정
df.describe()
```

```python
# 수치형 데이터 측정
df.describe(include=[np.number])
```

```python
# 범주형 데이터 측정
df.describe(include=['object'])
```

```python
# 특정 열 분석
df['score'].describe()
```

### 6-2 DataFrame 열 선택 및 **조작**

```python
import pandas as pd

data = {
    "이름": ["김철수", "이영희", "박민수", "최지훈", "정소희"],
    "학년": [1, 2, 3, 4, 2],
    "학점": [4.2, 3.8, 4.5, 3.9, 3.5],
    "학과": ["컴퓨터공학", "경영학", "전자공학", "의학", "심리학"],
    "동아리": ["프로그래밍", "독서토론", "로봇공학", "봉사활동", "음악감상"]
}
df = pd.DataFrame(data)
```

### 6-4 DataFrame 정렬

```python
import pandas as pd

data = {
    "이름": ["김철수", "이영희", "박민수", "최지훈", "정소희"],
    "학년": [1, 2, 3, 4, 2],
    "학점": [4.2, 3.8, 4.5, 3.9, 3.5],
    "학과": ["컴퓨터공학", "경영학", "전자공학", "의학", "심리학"],
    "동아리": ["프로그래밍", "독서토론", "로봇공학", "봉사활동", "음악감상"]
}

#DataFrame 인덱스 지정
df = pd.DataFrame(data).set_index('이름')
df
```

```python
#DataFrame 단일 열 오름차순, 내림차순 정렬
df.sort_values(by='학점') #오름차순
df.sort_values(by='학점', ascending=False) #내림차순
```

```python
#DataFrame 복수 열 오름차순, 내림차순 정렬
df.sort_values(by=['학년','학점'],ascending=[True,False]) #오름차순,내림차순
```

```python
#DataFrame 인덱스 기분 오른차순, 내림차순 정렬
df.sort_index(ascending=False) #내림차순
```
